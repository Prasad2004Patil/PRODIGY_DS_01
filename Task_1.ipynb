```python
import matplotlib.pyplot as plt
numeric_columns = [
    "Cases_Property_Recovered",
    "Cases_Property_Stolen",
    "Value_of_Property_Recovered",
    "Value_of_Property_Stolen"
]
for col in numeric_columns:
    plt.figure(figsize=(5, 5))
    plt.hist(df[col], color='#B22222', bins=10)
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert non-numeric columns to appropriate types if necessary
# You can use this to convert all columns to numeric (forcing errors to NaN, so they won't affect the sum)
df = df.apply(pd.to_numeric, errors='coerce')

# Group the data by 'Year' and sum numeric columns
grouped_data = df.groupby("Year").sum()

# Total values per year (sum of all numeric columns)
total_values = grouped_data.sum(axis=1)

# Plot the data
plt.figure(figsize=(10, 7))  # Adjusted size for readability
plt.barh(total_values.index.astype(str), total_values.values, color='#191970')
plt.xlabel('Total Values', size=14)
plt.ylabel('Year', size=14)
plt.title('Total Values per Year', size=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r'Task-1.csv')

# Ensure the column you want to plot exists and replace 'Task-1' with the correct column name
column_name = 'Cases_Property_Stolen'  # Replace this with the actual column name from your dataset
if column_name in df.columns:
    # Create a histogram
    df[column_name].plot(kind='hist', bins=20, color='blue', edgecolor='black')
    plt.title('Distribution of Cases Property Stolen')
    plt.xlabel('Cases Stolen')
    plt.ylabel('Frequency')
    plt.show()
else:
    print(f"Column '{column_name}' not found in the dataset.")
```

```python
file_path = 'Task-1.csv' 

if os.path.exists(file_path):
    data = pd.read_csv(file_path)
    print("File loaded successfully!")
else:
    print(f"File not found at: {file_path}")
```

```python
# Group by Area_Name and sum the cases
area_cases = data.groupby('Area_Name')['Cases_Property_Stolen'].sum()

# Create a bar chart
area_cases.plot(kind='bar', color='green')
plt.title('Total Cases Property Stolen by Area')
plt.xlabel('Area')
plt.ylabel('Total Cases Stolen')
plt.xticks(rotation=90)
plt.show()
```
